```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
total_iters = 100
x = np.linspace(0, total_iters-1, total_iters)
```

```python
lr = 0.02
min_lr = 0.002
rest_lr = min_lr + (lr - min_lr) * 0.5 * (1. + np.cos(np.pi * x / (total_iters-1)))
plt.plot(x, rest_lr)
```

```python
rest_lr
```

```python
lr = 0.002
max_lr = 0.02
pretrained_lr = max_lr + (lr - max_lr) * 0.5 * (1. + np.cos(np.pi * x / (total_iters-1)))
plt.plot(x, pretrained_lr)
```

```python
pretrained_lr[pretrained_lr > rest_lr] = rest_lr[pretrained_lr > rest_lr]
pretrained_lr
```

```python
plt.plot(x, rest_lr)
plt.plot(x, pretrained_lr)
plt.title('Learning Rate Schedule')
plt.xlabel('Training percentage'); plt.ylabel('Lr')
plt.legend(['rest params lr', 'pretrained params lr'])
```
